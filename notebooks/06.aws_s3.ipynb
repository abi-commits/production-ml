```python
import boto3
from pathlib import Path
```

```python
bucket = "housing-data-artifacts"
region = "ap-south-1"

PROJECT_ROOT = Path.cwd().parent
data_path = PROJECT_ROOT / "data" / "processed"
models_path = PROJECT_ROOT / "models"

s3 = boto3.client("s3", region_name=region)

def upload_to_s3(file_path: Path, s3_key: str):
    if not file_path.exists():
        print(f"File {file_path} does not exist.")
        return
    print(f"Uploading {file_path} to s3://{bucket}/{s3_key}")
    s3.upload_file(str(file_path), bucket, s3_key)

# Upload processed data files
upload_to_s3(data_path / "feature_engineered_train.csv", "data/processed/feature_engineered_train.csv")
upload_to_s3(data_path / "feature_engineered_eval.csv", "data/processed/feature_engineered_eval.csv")
upload_to_s3(data_path / "feature_engineered_holdout.csv", "data/processed/feature_engineered_holdout.csv")

# Upload model files
upload_to_s3(models_path / "xgb_model.pkl", "models/xgb_model.pkl")
```
